# Sesión 11: Función de Rosenbrock

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as opt
```

```python
def f(p):
    x = p[0]
    y = p[1]
    return (1-x)**2+100*(y-x**2)**2
```

```python
f([1.0,1.0])
```

```python
f([-1.0,1.0])
```

```python
x = np.linspace(-2, 2, 500)
y = np.linspace(-2, 2, 500)  
xm, ym = np.meshgrid(x, y)
z = (1-xm)**2 + 100*(ym-xm**2)**2
zz = np.log(z + 0.1)
```

```python
plt.contour(x, y, z, levels=150)
plt.colorbar()
```

```python
plt.contour(x, y, zz, levels=40)
plt.colorbar()
```

```python
params0 = np.array([0.0,0.0])
res = opt.minimize(f, params0, method='nelder-mead', tol=1e-9
                   ,options={'disp': True, 'maxiter' : 4000})
```

```python
res.x
```

```python
params0 = np.array([-1.5,1.5])
res = opt.minimize(f, params0, method='nelder-mead', tol=1e-9
                   ,options={'disp': True, 'maxiter' : 4000})
```

```python
res.x
```

```python
params0 = np.array([-1.5,1.5])
res = opt.minimize(f, params0, method='CG', tol=1e-9, jac='2-point'
                   ,options={'disp': True, 'maxiter' : 4000})
```

```python
res.x
```
